```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify=y)
```

```python
#wolking with tree model, not necessary to scaling the data , but if want to work with linear model it a must step
y.value_counts()
```

```python
from sklearn.utils.class_weight import compute_class_weight
classes =np.unique(y_train)
```

```python
classes
```

```python
weights =compute_class_weight(class_weight="balanced",classes=classes,y=y_train)
```

```python
weights
```

```python
class_weights= dict(zip(classes,weights))
```

```python
class_weights
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf=RandomForestClassifier(class_weight=class_weights)
```

```python
rf_params ={
    "n_estimators": [100,200,300],
    "max_depth":[5,10,20,None],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4],
}
```

```python
rf_random =RandomizedSearchCV(
    rf,
    rf_params,
    n_iter=20,
    cv=3,
    scoring="accuracy",
    n_jobs=-1
)
```

```python
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
```

```python
gb_params ={
    "n_estimators": [100,200],
    "learning_rate":[0.01 ,0.1 ,0.2],
    "max_depth":[3,5,7]
    
}
```

```python
gb_random = RandomizedSearchCV(
    gb,
    gb_params,
    n_iter= 15,
    cv=3,
    scoring="accuracy",
    n_jobs= -1
)
```

```python
gb_random.fit(X_train, y_train)
```

```python
from xgboost import XGBClassifier
xgb =XGBClassifier(
    scale_pos_weight=class_weights[0] / class_weights[1],
    eval_metric="logloss"
)
```

```python
xgb_params = {
    "n_estimators":[100 ,200,300],
    "max_depth":[3,5,7],
    "learning_rate":[0.01 , 0.1 , 0.2],
    "subsample":[0.7,1.0],
    "colsample_bytree":[0.7 ,0.1]
    
}
```

```python
xgb_random = RandomizedSearchCV(
    xgb,
    xgb_params,
    n_iter=20,
    cv=3,
    scoring="accuracy",
    n_jobs=-1
)
```

```python
xgb_random.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
models ={
    "RandomForest" :rf_random.best_estimator_,
    "GradientBoosting":gb_random.best_estimator_,
    "XGBoost":xgb_random.best_estimator_
}
```

```python
for name,model in models.items():
    print(f"{name}")
    y_pred = model.predict(X_test)
    print(accuracy_score(y_test,y_pred))
```
